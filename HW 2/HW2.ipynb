```python
###3a. 
expeditions=pd.read_csv("expeditions.csv")
merged_dataset= expeditions.merge(members, how='left', left_on="expedition_id", right_on="expedition_id")
merged_dataset.head()
```

```python
all_three_datasets= merged_dataset.merge(peaks, how='left', left_on="expedition_id", right_on="first_ascent_expedition_id")
non_hired=all_three_datasets.loc[all_three_datasets["hired"]==False] 
non_hired.info()
```

```python
non_hired_women=non_hired[(non_hired["sex"]=="F")&(non_hired["success"]==True)]
non_hired_women.info()
```

```python
earliest_women= non_hired_women.sort_values(by=["year_x"])
print(earliest_women)
```

# 4A) 
The number of expeditions increased throughout the decades. Starting in 1980, the number of expeditions increased drastically and nearly doubled from 1990 to 2000. 

I chose to look at the total number of expedition members using the 'members' variable, because I am assuming expedition members means members and not hired staff as well. The total number of members also increased throughout the decades. There was a decrease from 1930 to 1940, which I am assuming is because of WWII. The number of expedition members took a large jump between 1960 to 1970, and then after that began to drastically increase again with almost 3 times as many members from 1970 to 1980. 

The averate rates of member deaths does not get higher than 40% in 1920. There were no deaths in 1910 and 1940. There is no consistent pattern of average member deaths throughout the decades. 

The average rates of death for hired staff is lower than the average rates of member deaths. For the two decades (1920 and 1900) where there are 100% average rates of death, this is because the death rate is set to 100% if the number of hired deaths is greater than or equal to the number of hired staff.
