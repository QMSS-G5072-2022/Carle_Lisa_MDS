## a) Length of Narratives

```python
###Read in the data
import pandas as pd
import re
emergency=pd.read_csv('./data/emergency.csv', header = None)
emergency.columns = ['Description']
descriptions=emergency.iloc[:,0]
```

```python
###Finding longest and shortest description
description_lengths=[len(scripts) for scripts in descriptions]
max_length=max(description_lengths)
print(max_length)
min_length=min(description_lengths)
print(min_length)
max_length_rows=[]
min_length_rows=[]

for rows in descriptions:
    if len(rows)== max_length:
        max_length_rows.append(rows)
    if len(rows)== min_length:
        min_length_rows.append(rows)
```

```python
### Printing longest descriptions
print("The longest descriptions are:\n"+"\n".join([rows for rows in max_length_rows]))
```

```python
### Printing shortest descriptions
print("The shortest descriptions are:\n" +"\n".join([rows for rows in min_length_rows]))
```

## b) Fractures



```python
###Finding all narratives related to "fractures", Counting how many fracture narratives I found
fracture_finder=[narratives for narratives in descriptions if re.search(r'\b(FRACTUR|FX)',narratives)] 
print('There are {} narratives related to "fractures"'.format(len(fracture_finder)))
```

## c) Body Parts

```python
###Identifying the body parts that are fractured
bodypartdict={}
def dictfunction(bodypart):
    if bodypart[0] in bodypartdict:
        bodypartdict[bodypart[0]]+=1
    else:
        bodypartdict[bodypart[0]]=1

for narrative in fracture_finder:
    body_part1=re.findall(r'FRACTUR\w+\s(SKULL|SHOULDERS?|\bKNEES?|LEG)',narrative)
    if body_part1!=[]:
        dictfunction(body_part1)
    body_part2=re.findall(r'(SKULL|SHOULDERS?|\bKNEES?|LEG)\sFRACTUR',narrative)
    if body_part2!=[]:
        dictfunction(body_part2)
    body_part3=re.findall(r'FX.*\s(SKULL|SHOULDERS?|\bKNEES?|LEG)',narrative)
    if body_part3!=[]:
        dictfunction(body_part3)
    body_part4=re.findall(r'(SKULL|SHOULDERS?|\bKNEES?|LEG)\sFX.*',narrative)
    if body_part4!=[]:
        dictfunction(body_part4)

print(bodypartdict)
```

```python
#making histogram
import matplotlib.pyplot as plt

plt.bar(bodypartdict.keys(), bodypartdict.values())
plt.show()
```

# d) Age and Gender

```python
###Adding new columns
emergency= emergency.assign(new_variables=[age for age in [re.findall(r'(\d{1,2})\s*(YO|Y|YR|Y|MO|MONTH\sOLD)O*\s*(F|M|FEMALE|MALE)*,*-*\s*', rows) for rows in emergency['Description']]])
header_list= ['Description', 'new_variables', 'Age','Gender']
emergency= emergency.reindex(columns = header_list)

for index,rows in emergency.iterrows():
    for values in rows['new_variables']:
        if values[1]=="MO" or values[1]=="MONTH OLD":
            emergency.iloc[index,2]= int(values[0])/12
        if values[1]== "Y" or values[1]=="YO" or values[1]=="YR":
            emergency.iloc[index,2]= int(values[0])
        if values[2]=="FEMALE"or values[2]=="F":
            emergency.iloc[index,3]='FEMALE'
        if values[2]=="MALE" or values[2]=="M":
            emergency.iloc[index,3]='MALE'
        
            
            
```

```python
###How many men and women in data
emergency.groupby('Gender').count()
```

```python
###Graphical representation of the number of injuries vs age disaggregated by gender
emergency.hist(column="Age", by = "Gender", bins=10)
```
