```python
### a) Left join the trial and confession tables.For what share of trials does the database record confessions? Create a results table with the number of all trials, the number of confessions, and the share of trials with confessions recorded.
```

```python
%%sql 
SELECT COUNT(t.trialref) AS 'Number of Trials',
COUNT(c.trialref) AS 'Number of Confessions',
(COUNT(c.trialref)/COUNT(t.trialref) * 100) AS 'Share of Trials w Confessions Recorded'
FROM trial as t left join confession as c
on t.trialref = c.trialref
```

```python
### b) Only a small number of trials have records of torture. Is there a higher share of confessions among trials with records of torture than trials without such record? Hint: You will need to merge on the confession table.
```

```python
%%sql
SELECT (COUNT(c.trialref)/COUNT(t.trialref)*100) AS 'Proportion of Confessions among Trials with Torture'
FROM trial as t left join confession as c
on t.trialref = c.trialref
WHERE t.pretrialnotes LIKE '%torture%'
```

```python
%%sql
SELECT (COUNT(c.trialref)/COUNT(t.trialref)*100) AS 'Proportion of Confessions among Trials without Torture'
FROM trial as t left join confession as c
on t.trialref = c.trialref
WHERE t.pretrialnotes NOT LIKE '%torture%'
```
